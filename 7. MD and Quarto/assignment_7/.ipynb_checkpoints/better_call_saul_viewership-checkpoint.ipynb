```python
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
```

```python
ratings = pd.read_csv("vievership_stats.csv")

# Show title and table without index
display(HTML("<h3>Better Call Saul Season Viewership</h3>"))
display(HTML(ratings.to_html(index=False)))
```

```python
#| label: fig-viewership-trend
#| fig-cap: "Average Viewership per Season for Better Call Saul"

# Plot
plt.figure(figsize=(10, 5))
plt.plot(ratings["Season"], ratings["Average Viewers"], marker='o', linestyle='-', linewidth=2)

plt.title("Average Viewership per Season for Better Call Saul")
plt.xlabel("Season")
plt.ylabel("Average Viewers (millions)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#| label: fig-viewership-change
#| fig-cap: "Change in Average Viewership Between Seasons"

# Calculate change in average viewership between seasons
ratings["Change"] = ratings["Average Viewers"].diff()

# Plot
plt.figure(figsize=(10, 5))
plt.bar(ratings["Season"][1:], ratings["Change"][1:], color='lightblue', edgecolor='black')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

plt.title("Change in Average Viewership Between Seasons")
plt.xlabel("Season")
plt.ylabel("Change in Viewers (millions)")
plt.tight_layout()
plt.show()
```

```python
# Ensure season is a string for sorting and display
ratings["Season"] = ratings["Season"].astype(str)

# Calculate change from previous season
ratings["Change"] = ratings["Average Viewers"].diff()

# Create a clean summary table, skip first season (no previous data)
change_table = ratings[["Season", "Change"]].copy()
change_table = change_table[1:]  # Skip the first row
change_table.columns = ["Season", "Change since last season (in milions)"]

# Round for readability
change_table["Change since last season (in milions)"] = change_table["Change since last season (in milions)"].round(2)

# Display with a title
display(HTML("<h3>Average Viewership Changes</h3>"))
display(HTML(change_table.to_html(index=False)))
```
